##### Price e autos apresentam valores do tipo string.Logo, serÃ¡ necessÃ¡rio converter esses valores para int

```python
# vamos converter esses dados atravÃ©s do astype()
# converter o tipo de dados de um objeto em outro tipo de dados especificado. 
# Geralmente, Ã© usado com objetos
# do tipo numpy.ndarray para realizar conversÃµes de tipo em arrays.
# str.replace() Ã© usado para substituir ocorrÃªncias de um determinado valor em uma string por outro valor especificado
autos["price"] = (autos['price'].astype(str).str.replace("$", '').str.replace(',', ",").astype(int))
autos['price'].head()
```

```python
autos['odometer'] = (autos['odometer'].astype(str).str.replace('km', '').str.replace(',', ',').astype(int))
autos['odometer'].head()
```

```python
autos.rename({'odometer': 'odometer_km'}, axis=1, inplace=True)
autos['odometer_km'].head()
```

```python

(autos["date_Crawled"]
        .str[:10]
        .value_counts(normalize=True, dropna=False)
        .sort_values()
        *100)
```

```python
(autos["lastSeen"]
        .str[:10]
        .value_counts(normalize=True, dropna=False)
        .sort_values()
        *100)
```

normalize=True: Quando definido como True, o mÃ©todo value_counts() retorna a contagem de cada valor normalizada, ou seja, a proporÃ§Ã£o de cada valor em relaÃ§Ã£o ao nÃºmero total de valores. Os valores normalizados sÃ£o expressos como fraÃ§Ãµes decimais ou porcentagens, dependendo da configuraÃ§Ã£o do pandas.

dropna=True: Quando definido como True, o mÃ©todo value_counts() exclui os valores ausentes (NaN) da contagem. Isso significa que os valores NaN nÃ£o serÃ£o considerados no resultado.
