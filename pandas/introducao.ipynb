Análise de dados aprimorada : ao usar valores NaN para representar dados ausentes, o restante dos dados ainda pode ser analisado e processado, permitindo uma melhor compreensão dos padrões e relacionamentos nos dados.

Legibilidade aprimorada : conjuntos de dados que usam valores NaN para representar dados ausentes são mais legíveis, pois indica claramente a presença de valores ausentes no conjunto de dados.

Representação fácil de dados ausentes : ao usar valores NaN, os dados ausentes em um conjunto de dados podem ser facilmente representados e identificados, em vez de usar valores de espaço reservado, como 0ou -1que podem dificultar ou enganar a interpretação dos resultados dos cálculos.

O índice da série que Series.value_counts()produz agora nos mostra floats como 471.0em vez de inteiros. A razão por trás disso é que os pandas usam o tipo inteiro NumPy, que não suporta valores NaN. Os pandas herdam esse comportamento e, nos casos em que você tenta atribuir um valor NaN a uma coluna inteira, os pandas convertem silenciosamente essa coluna em um float dtype
